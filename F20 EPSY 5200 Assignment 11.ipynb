```python
# Problem 1 code here (can split into multiple code chunks if you want)
import pandas as pd

school_data = pd.read_csv('MA_Public_Schools_2017.csv')
print(school_data)

# starter code for 1c -- replace school_data w/ your dataframe name
school_data.columns = school_data.columns.str.replace(' ', '_') # notice this replaces the 1st argument w/ the 2nd
print(school_data.columns)

# replace % with Perc
school_data.columns = school_data.columns.str.replace('%','Perc')
print(school_data.columns)


# subset columns with numeric data
num_school_data = (school_data[['PK_Enrollment','K_Enrollment',
       '1_Enrollment', '2_Enrollment', '3_Enrollment', '4_Enrollment',
       '5_Enrollment', '6_Enrollment', '7_Enrollment', '8_Enrollment',
       '9_Enrollment', '10_Enrollment', '11_Enrollment', '12_Enrollment',
       'SP_Enrollment', 'Perc_First_Language_Not_English',
       'Perc_English_Language_Learner', 'Perc_Students_With_Disabilities',
       'Perc_High_Needs', 'Perc_Economically_Disadvantaged',
       'Perc_African_American', 'Perc_Asian', 'Perc_Hispanic', 'Perc_White',
       'Perc_Native_American', 'Perc_Native_Hawaiian,_Pacific_Islander',
       'Perc_Multi-Race,_Non-Hispanic', 'Perc_Males', 'Perc_Females',
       'Average_Class_Size', 'Number_of_Students', 'Salary_Totals',
       'Average_Salary', 'FTE_Count', 'In-District_Expenditures',
       'Total_In-district_FTEs', 'Average_In-District_Expenditures_per_Pupil',
       'Total_Expenditures', 'Total_Pupil_FTEs',
       'Average_Expenditures_per_Pupil']])
print(num_school_data)

# descriptive statistics for numeric data
desc_school_data = (num_school_data).describe().transpose()
print(desc_school_data)

#finished problem 1
```

These plots are a little hard for me to interpret. I think schools included largely have class sizes around 20 students, most of the teachers have a salary at or above $60k, the percentage of students who are economically disadvantaged seems relatively low and there aren't that many English learners.

```python
# post hoc exploratory analysis looking at relationship between teacher salary and class size
sns.regplot(school_data2.Average_Salary , school_data2.Average_Class_Size)

lm = smf.ols(formula = "Average_Class_Size ~ Average_Salary", data = school_data2).fit()
lm.summary()
```

The R-squared value is low and the model only accounts for 0.9% of the variance in the dependent variable (class size). It looks like teacher salary can predict class size--there is a significant positive relationship between salary and class size. I was assuming that teachers with more seniority would make more money and be able to have smaller class sizes.
